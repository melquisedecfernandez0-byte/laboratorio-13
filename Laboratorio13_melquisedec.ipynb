Prompt 1: Análisis de Sentimiento con TextBlob

Instrucciones:
Utiliza la librería TextBlob para analizar el sentimiento de las siguientes frases. Debes interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su polaridad.

Frase 1: "The new turbine cooling system shows excellent efficiency and outstanding thermal performance."
Frase 2: "The fatigue test resulted in a severe structural failure due to poor material selection."
Frase 3: "The engineering team is reviewing the technical specifications of the hydraulic pump."
Objetivo:
Para cada frase, analiza la polaridad y asigna el sentimiento correspondiente:

Positivo si la polaridad es mayor que 0.
Negativo si la polaridad es menor que 0.
Neutral si la polaridad es 0.

```python
from textblob import TextBlob

# Definición de las frases de ingeniería mecánica
frases = [
    "The new turbine cooling system shows excellent efficiency and outstanding thermal performance.",
    "The fatigue test resulted in a severe structural failure due to poor material selection.",
    "The engineering team is reviewing the technical specifications of the hydraulic pump."
]

def analizar_sentimiento_ingenieria(lista_frases):
    print(f"{'FRASE':<90} | {'POLARIDAD':<10} | {'SENTIMIENTO'}")
    print("-" * 120)

    for frase in lista_frases:
        # Procesamiento con TextBlob
        analisis = TextBlob(frase)

        # Obtención de la polaridad (-1.0 a 1.0)
        # La polaridad mide qué tan positiva o negativa es la frase
        polaridad = analisis.sentiment.polarity

        # Lógica de asignación según el objetivo
        if polaridad > 0:
            sentimiento = "Positivo"
        elif polaridad < 0:
            sentimiento = "Negativo"
        else:
            sentimiento = "Neutral"

        # Imprimir resultados formateados
        print(f"{frase:<90} | {polaridad:<10.2f} | {sentimiento}")

# Ejecución del script
if __name__ == "__main__":
    analizar_sentimiento_ingenieria(frases)
```

Frase 1: "The new turbine cooling system shows excellent efficiency and outstanding thermal performance."
Traducción: "El nuevo sistema de refrigeración de la turbina muestra una eficiencia excelente y un rendimiento térmico sobresaliente."
Frase 2: "The fatigue test resulted in a severe structural failure due to poor material selection."
Traducción: "La prueba de fatiga resultó en un fallo estructural grave debido a una mala selección de materiales."
Frase 3: "The engineering team is reviewing the technical specifications of the hydraulic pump."
Traducción: ""El equipo de ingeniería está revisando las especificaciones técnicas de la bomba hidráulica."

Prompt 2: Entrenamiento de Modelo de Clasificación de Sentimientos con Naive Bayes

Instrucciones:
Entrena un modelo de clasificación de sentimientos utilizando el algoritmo Naive Bayes. Utiliza los siguientes ejemplos de frases etiquetadas y evalúa la precisión de tu modelo.

Datos de ejemplo:

Frase 1: "This app is very intuitive!"
Frase 2: "I didn't enjoy the customer service."
Frase 3: "Quite average experience."
Frase 4: "Fantastic presentation!"
Frase 5: "The food was awful."

Objetivo:

Entrena un modelo de Naive Bayes con estas frases etiquetadas.
Evalúa la precisión del modelo utilizando las frases proporcionadas.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# 1. Datos de ejemplo y sus etiquetas (0=Negativo, 1=Neutral, 2=Positivo)
frases = [
    "This app is very intuitive!",          # Positivo
    "I didn't enjoy the customer service.", # Negativo
    "Quite average experience.",            # Neutral
    "Fantastic presentation!",              # Positivo
    "The food was awful."                   # Negativo
]

# Etiquetas correspondientes
# Positivo: 2, Neutral: 1, Negativo: 0
etiquetas = [2, 0, 1, 2, 0]

# 2. Vectorización: Convertir texto a una matriz de conteo de palabras
vectorizador = CountVectorizer()
X = vectorizador.fit_transform(frases)

# 3. Entrenamiento del modelo Naive Bayes (Multinomial)
modelo = MultinomialNB()
modelo.fit(X, etiquetas)

# 4. Evaluación del modelo
# En este caso, evaluamos con los mismos datos de entrenamiento
predicciones = modelo.predict(X)
precision = accuracy_score(etiquetas, predicciones)

# 5. Mostrar resultados
print(f"Precisión del modelo: {precision * 100}%")
print("-" * 30)

mapping = {0: "Negativo", 1: "Neutral", 2: "Positivo"}
for frase, pred in zip(frases, predicciones):
    print(f"Frase: '{frase}' -> Predicción: {mapping[pred]}")
```

**Frase 1:** "This app is very intuitive!"  
**Traducción:** "¡Esta aplicación es muy intuitiva!"  
  
**Frase 2:** "I didn't enjoy the customer service."  
**Traducción:** "No disfruté del servicio al cliente."  
  
**Frase 3:** "Quite average experience."  
**Traducción:** "Experiencia bastante promedio."  
  
**Frase 4:** "Fantastic presentation!"  
**Traducción:**  "¡Presentación fantástica!"  
  
**Frase 5:** "The food was awful."  
**Traducción:** "La comida estaba horrible."

```python
nueva_frase = ["this food is very good"]
nueva_vectorizada = vectorizador.transform(nueva_frase)

# Imprimir la representación vectorizada para entender por qué la clasificación podría ser incorrecta
# Si la frase contiene palabras que no están en el vocabulario del modelo, el vector será escaso (principalmente ceros).
# En este caso, 'i', 'am', 'happy' no están en el vocabulario del modelo entrenado.
print(f"Vectorizada de '{nueva_frase[0]}': {nueva_vectorizada.toarray()}")
print("Nota: Este vector es escaso (principalmente ceros) porque las palabras 'i', 'am', 'happy' no se encontraron en el vocabulario de entrenamiento del modelo.")
print("Cuando un modelo Naive Bayes recibe un vector de entrada con solo ceros, a menudo predice la clase con la probabilidad previa más alta o una clase predeterminada, que en este caso resulta ser 'Negativo'.")

resultado = modelo.predict(nueva_vectorizada)
print(f"Nueva frase: {mapping[resultado[0]]}")
```
